```python
import pandas as pd
```

```python
df = pd.read_html("https://areacode.org/")[0]

df = df.drop("Area Code.1", axis=1)

df = df.rename(columns={
    "Area Code": "area_code",
    "State/Province": "state_province",
    "Major Cities By Population": "cities",
})

df["area_code"] = df["area_code"].str.extract(r"(\d+)")
```

```python
df.to_json("out/codes.json", orient="records")
```
